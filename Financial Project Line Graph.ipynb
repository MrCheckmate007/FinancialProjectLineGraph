```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1. **Draw a Line Plot**

```python
x = [34, 83, 92, 100, 101, 110]
y = [12, 50, 88, 73, 92, 112]
plt.plot(x,y)
plt.show()
```

## 2.  **Insert Title, X-Label and Y-Label**

```python
x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
y = [34, 83, 92, 100, 101, 110]
plt.plot(x,y)
plt.title("Financial Project Line Plot")
plt.xlabel("2020-Months")
plt.ylabel("Price")
plt.show
```

## 3. **Change Font Size of the X and Y Labels and their locations**

```python
x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
y = [34, 83, 92, 100, 101, 110]
plt.plot(x,y)
plt.title("Financial Project Line Graph", fontsize = 20, loc="left")
plt.xlabel("2020-Months", fontsize = 20, loc ='right')
plt.ylabel("Price", fontsize = 20, loc = 'bottom')
plt.show()
```

## 4. **Changing Color and Line Width**

```python
x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
y = [34, 83, 92, 100, 101, 110] 
plt.plot(x,y, color = 'green', linewidth = 5)
plt.title("Financial Project Line Graph", fontsize = 16, loc = "left")
plt.xlabel("2020-Months", fontsize = 16, loc = "right")
plt.ylabel("Price", fontsize = 16, loc = "top")
plt.show()
```

## 5. **Marker, Marker Size and Line Style

```python
x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
y = [34, 83, 92, 100, 101, 110] 
plt.plot(x,y,'g--.', linewidth=3, markersize = 15, 
        markerfacecolor='black', markeredgecolor='red')
plt.title("Financial Project Line Graph", fontsize = 14, loc = 'right')
plt.xlabel("2020-Month", fontsize=14, loc= "right")
plt.ylabel("Price", fontsize = 14, loc = 'bottom')
plt.show()
```

## 6. **Adding a Legend to the Graph and Chaning the Location**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x,y, 'g--.', linewidth=3, markersize=10, 
        markerfacecolor='black', markeredgecolor='red')
plt.title("Financial Project Line Graph", fontsize=14, loc='right')
plt.xlabel("X-Axis", fontsize=14, loc='left')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.legend(["Simple Line Plot"], loc=0)
plt.show()
```

## 7. **Create Seperate Lines and styles for x and y**

```python
plt.style.use('seaborn')
```

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.legend()
plt.show()
```

## 8. **X lim and Y lim**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.xlim(0,20)
plt.ylim(0,100)
plt.legend()
plt.show()
```

## 9. **Change X and Y limits Using the axis method**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.axis([0,20,0,100])
plt.legend()
plt.show()
```

## 10. **Change X and Y limit Using Xticks and Yticks**

```python
x_ticks = [x for x in range (0,20,4)]
y_ticks = [y for y in range (0,100,10)]
```

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

## 11. **Using the Grid Method**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='black', linestyle='-.', linewidth=1)
plt.show()
```

## 12. **Change Figure Size**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.figure(figsize=(12,5), facecolor='lightgreen', edgecolor='black', linewidth = 2)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='black', linestyle='-.', linewidth=1)
plt.show()
```

## 13. **Saving the Graph**

```python
x = np.random.randint(low=10,high=100,size=20)
y = np.random.randint(low=10,high=100,size=20)
plt.figure(figsize=(12,5), facecolor='lightgreen', edgecolor='black', linewidth = 2)
plt.plot(x, "g--.", linewidth=3, markersize = 10, 
        markerfacecolor="black", markeredgecolor='Blue', label="Line 1")
plt.plot(y, "b", linewidth=3, markersize = 10, marker = '.',
        markerfacecolor ="red", markeredgecolor="black", label="Line 2")
plt.title("Financial Project Line Graph", fontsize=14, loc= 'left')
plt.xlabel("X-Axis", fontsize=14, loc='right')
plt.ylabel("Y-Axis", fontsize=14, loc='top')
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.grid(color='black', linestyle='-.', linewidth=1)
plt.savefig('Financial Project Line Graph.png')
plt.show()
```
