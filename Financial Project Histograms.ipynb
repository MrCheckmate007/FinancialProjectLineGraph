## **Importing Libraries**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('C:/Users/Micha/titanic.csv')
data.head()
```

## 1. **Plotting a Histogram for the Age Column**

```python
data.columns
```

```python
plt.hist(data['Age'])
plt.title('Financial Project Histogram')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 2. **Bins Parameter Method**

```python
data["Age"].value_counts(bins=10)
```

```python
plt.hist(data["Age"], bins=[10,15,20,30,40,50,60,70,80])
plt.title('Financial Project Histogram')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 3. **Set the Width between the Bars**

```python
plt.hist(data['Age'], bins =[10,15,20,30,40,50,60,70,80],
        rwidth=0.80)
plt.title("Financial Project Histogram")
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

## 4. **Change Histtype and Align the Bars to the Middle**

```python
plt.hist(data['Age'], bins = [10,15,20,30,40,50,60,70,80],
        rwidth = 0.80, histtype='barstacked', align = 'mid')
plt.title("Financial Project Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 5. **Change Orientation to Horizontal**

```python
plt.hist(data['Age'], bins=[10,15,20,30,40,50,60,70,80],
        rwidth=0.80, histtype='barstacked', align='mid', orientation = 'horizontal')
plt.title("Financial Histogram Project")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(data["Age"], bins=[10,15,20,30,40,50,60,70,80],
        rwidth=0.80, histtype='barstacked', align ='mid',
        orientation = 'vertical', color = "g")
plt.title('Financial Histogram Project')
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## 7. **Draw Histograms for Two data sets**

```python
Subject1_scores= np.random.randint(low=1,high=20,size=15)
Subject2_scores= np.random.randint(low=1,high=20,size=15)
```

```python
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['r','b'], 
        label = ["Subject 1 Scores", "Subject 2 Scores"])
plt.title("Financial Histogram Project")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

## 8. **Change Limit of X and Y Axis**

```python
x_ticks = [x for x in range(1,20)]
y_ticks = [x for x in range(1,10)]
```

```python
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['r','b'], 
        label = ["Subject 1 Scores", "Subject 2 Scores"])
plt.title("Financial Histogram Project")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

## 9. **Change Figure Size & Color**

```python
plt.figure(figsize=(10,3), facecolor = 'lightblue', 
           edgecolor = 'black', linewidth = 3)
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['r','b'], 
        label = ["Subject 1 Scores", "Subject 2 Scores"])
plt.title("Financial Histogram Project")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5), facecolor = 'lightblue', 
           edgecolor = 'black', linewidth = 3)
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['r','b'], 
        label = ["Subject 1 Scores", "Subject 2 Scores"])
plt.title("Financial Histogram Project")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.xticks(x_ticks)
plt.yticks(y_ticks)
plt.legend()
plt.show()
```

## 11. **Changing Wording size and location**

```python
plt.figure(figsize=(10,5), facecolor = 'lightblue', 
           edgecolor = 'black', linewidth = 3)
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['g','b'],
        label = ["Subject 1 Scores", "Subject 2 Scores"])
#label = ["Subject 1 Scores", "Subject 2 Scores"]
plt.title("Financial Histogram Project", fontsize = 22, loc = 'center', style = 'italic')
plt.xlabel("Age", fontsize = 14, loc = 'left')
plt.ylabel("Frequency", fontsize = 14, loc = 'top')
plt.xticks(x_ticks, rotation = 50, style='oblique')
plt.yticks(y_ticks, style = 'oblique')
plt.legend(loc= 'upper right', bbox_to_anchor=(0.5,0.5, 0.75, 0.5), frameon=True)
plt.grid(color = 'grey', linestyle = '--', linewidth = 1)
plt.show()
```

## 12. **Saving the File**

```python
plt.figure(figsize=(10,5), facecolor = 'lightblue', 
           edgecolor = 'black', linewidth = 3)
plt.hist([Subject1_scores, Subject2_scores], bins=5, rwidth=0.80, histtype = "bar",
        align = 'mid', orientation = 'vertical', color = ['g','b'],
        label = ["Subject 1 Scores", "Subject 2 Scores"])
#label = ["Subject 1 Scores", "Subject 2 Scores"]
plt.title("Financial Histogram Project", fontsize = 22, loc = 'center', style = 'italic')
plt.xlabel("Age", fontsize = 14, loc = 'left')
plt.ylabel("Frequency", fontsize = 14, loc = 'top')
plt.xticks(x_ticks, rotation = 50, style='oblique')
plt.yticks(y_ticks, style = 'oblique')
plt.legend(loc= 'upper right', bbox_to_anchor=(0.5,0.5, 0.75, 0.5), frameon=True)
plt.grid(color = 'grey', linestyle = '--', linewidth = 1)
plt.savefig('Financial Histogram Project')
plt.show()
```
