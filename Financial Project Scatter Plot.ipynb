```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
temp= [14.2,16.4,11.9,15.2,18.5,22.1,19.4]
icecream_sales=[215,325,185,332,406,522,412]
tea_sales=[80,65,89,70,56,45,43]
```

## 1. **Draw The Scatter Plot with Title, X and Y Label**

```python
plt.scatter(temp, icecream_sales)
plt.title("Temp Vs. Ice Cream Sales")
plt.xlabel("Temp")
plt.ylabel("Ice Cream Sales")
plt.show()
```

## 2. **Change Size of The-Dots**

```python
plt.scatter(temp, icecream_sales, s =100)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 3. **Change the Color of the Dots**

```python
plt.scatter(temp, icecream_sales, s =100, c = 'g')
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 4. **Change The Marker**

```python
plt.scatter(temp, icecream_sales, s =100, c='g'
           marker = 'd')
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

```python
plt.scatter(temp, icecream_sales, s =100, c = 'g',
           marker ="d")
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 5. **Changing the Edge Color of The Marker and Line Width**

```python
plt.scatter(temp, icecream_sales, s =100, c = 'g',
           marker ="d", edgecolor= 'b',
           linewidth = 2)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 6. **Change Alpha Parameter**

```python
plt.scatter(temp, icecream_sales, s =100, c = 'g',
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.25)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 7. **Set Different Colors For Different Dots**

```python
color =[x for x in range(11,18)]
color
```

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 8. **Use Built-In Map Colors**

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75, cmap="Oranges")
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.show()
```

## 9. **Add The ColorBar**

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75, cmap="Oranges")
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
plt.colorbar()
plt.show()
```

## 10. **Customize The Limit of Colorbar**

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75, cmap="Oranges",
           vmin = 10, vmax=20)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
c_bar =plt.colorbar()
c_bar.set_label("Temp Vs. Ice Cream Sales")
plt.show()
```

## 11. **Set Different Size for Different Dots**

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=1, cmap="Greens",
           vmin = 10, vmax=20)
plt.scatter(temp, tea_sales, s =100, c = color,
           marker ="d", edgecolor= 'black',
           linewidth = 2, alpha=1, cmap="Reds",
           vmin = 10, vmax=20)
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
c_bar =plt.colorbar()
c_bar.set_label("Temp Vs. Ice Cream Sales")
plt.show()
```

## 13. **Add Legend To the Plot**

```python
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75, cmap="Greens",
           vmin = 10, vmax=20, label = "Ice Cream Sales")
plt.scatter(temp, tea_sales, s =100, c = color,
           marker ="d", edgecolor= 'black',
           linewidth = 2, alpha=1, cmap="Reds",
           vmin = 10, vmax=20, label = "Tea Sales")
plt.title("Temp Vs. Ice Cream Sales", fontsize = 20, color = 'Red')
plt.xlabel("Temp", fontsize =14, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 14, color = 'b')
c_bar =plt.colorbar()
c_bar.set_label("Temp Vs. Ice Cream Sales")
plt.legend()
plt.show()
```

## 14. **Change Figure Size and Color**

```python
plt.figure(figsize=(12,5), facecolor="lightblue", edgecolor='yellow', linewidth=3)
plt.scatter(temp, icecream_sales, s =100, c = color,
           marker ="d", edgecolor= 'b',
           linewidth = 2, alpha=0.75, cmap="Greens",
           vmin = 10, vmax=20, label = "Ice Cream Sales")
plt.scatter(temp, tea_sales, s =100, c = color,
           marker ="d", edgecolor= 'black',
           linewidth = 2, alpha=1, cmap="Reds",
           vmin = 10, vmax=20, label = "Tea Sales")
plt.title("Temp Vs. Ice Cream Sales", fontsize = 24, color = 'Red',
         style ='italic')
plt.xlabel("Temp", fontsize =16, color = 'purple')
plt.ylabel("Ice Cream Sales", fontsize = 16, color = 'b')
c_bar =plt.colorbar()
c_bar.set_label("Temp Vs. Ice Cream Sales")
plt.legend()
plt.show()
```
