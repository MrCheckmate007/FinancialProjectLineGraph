## **Visulization of Data**

```python
# Plot Distribution
sns.displot(df['tip'])
```

```python
sns.displot(df['total_bill'], color='g')
```

```python
## Pairplot
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex', palette = 'husl')
```

```python
sns.pairplot(df, hue='smoker', palette = 'husl')
```

```python
## Bar Plots
sns.barplot(x='total_bill', y='sex', data =df, palette = 'coolwarm')
```

```python
sns.barplot(x='day', y='tip', hue ='smoker', data=df, palette = 'magma')
```

```python
## Countplot

df['sex'].value_counts()
```

```python
sns.countplot('sex', data=df, palette = 'muted')
```

```python
df['day'].value_counts()
```

```python
sns.countplot('day', data=df, palette='hls', hue='smoker')
```

```python
## violin plots
sns.violinplot(x='sex',y='total_bill', data=df, palette='rocket')
```

```python
sns.violinplot(x='total_bill', y='day', data=df, hue='time', palette='Accent')
```

```python
## Box Plots
sns.boxplot('smoker', 'total_bill', data=df, palette='husl')
```

```python
# joint plots
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

```python
## Scatter Plots
sns.catplot(x='day', y='tip', data=df)
```

```python
sns.catplot(x='sex', y='tip', data=df, jitter=False)
```

```python
sns.catplot(x='smoker', y='tip', kind = 'swarm', data=df, hue='sex')
```

```python
sns.heatmap(df.corr(), annot=True)
```
