```python
movie = ['Comedy','Action','Romance','Drama']
votes=[5,6,3,1]
```

```python
import matplotlib.pyplot as plt
import pandas as np
import numpy as pd
```

## 1. **Draw A Pie Chart**

```python
plt.pie(votes)
plt.show()
```

## 2. **Give Title, Label and Label Distance and  Legend**

```python
plt.title("Favorite Type of Movie")
plt.pie(votes, labels = movie, labeldistance=1.1)
plt.legend(loc = 'upper right')
plt.show()
```

## 3. **Pull out Action and Drama Wedges**

```python
plt.title("Favorite Type of Movie")
plt.pie(votes, labels = movie, labeldistance=1.1,
       explode = [0,0,0,0])
plt.legend(loc = 4)
plt.show()
```

## 4. **Change the Start Angle to 90 Degree**

```python
plt.title("Favorite Type of Movie")
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0.05],
       startangle = 90)
plt.legend(loc=4)
plt.show()
```

## 5.**Set The Color of Each Wedge**

```python
plt.title("Favorite Type of Movie")
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 90, 
        colors = ['g','red','pink','b'] )
plt.legend(loc=4)
plt.show()
```

## 6. **Display Percentage & Change the Distance inside of Each Wedge**

```python
plt.title("Favorite Type of Movie")
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 0, 
        colors = ['g','red','pink','b'],
       autopct = "%0.2f%%",
       pctdistance=0.6)
plt.legend(loc=4)
plt.show()
```

## 7. **Change the Radius of The Pie Chart**

```python
plt.title("Favorite Type of Movie", y=1.4)
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 0, 
        colors = ['g','red','pink','b'],
       autopct = "%0.2f%%",
       pctdistance=0.6,
       radius = 2)
plt.legend(loc=4)
plt.show()
```

## 8. **Set Labels Size of The Pie Chart**

```python
plt.title("Favorite Type of Movie", y=1.4)
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 0, 
        colors = ['g','red','pink','b'],
       autopct = "%0.2f%%",
       pctdistance=0.6,
       radius = 2,
       textprops={'fontsize' : 12})
plt.legend(loc=4)
plt.show()
```

## 9. **Set Border of the Ridges**

```python
plt.title("Favorite Type of Movie", y=1.4)
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 0, 
        colors = ['g','red','pink','b'],
       autopct = "%0.2f%%",
       pctdistance=0.6,
       radius = 2,
       textprops={'fontsize' : 12},
       wedgeprops = {'edgecolor': 'purple', 
                     'linewidth' : 3, 'linestyle' : ':'})
plt.legend(loc=4)
plt.show()
```

## 10. **Add Shadow to The Pie Chart**

```python
plt.title("Favorite Type of Movie", y=1.4)
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 0, 
        colors = ['g','red','pink','b'],
       autopct = "%0.2f%%",
       pctdistance=0.6,
       radius = 2,
       textprops={'fontsize' : 12},
       wedgeprops = {'edgecolor': 'purple', 
                     'linewidth' : 3, 'linestyle' : ':'},
       shadow = True)
plt.legend(loc=4)
plt.show()
```

## 11. **Change Figure Size and Color**

```python
plt.figure(figsize=(12,7), facecolor='lightyellow', edgecolor = 'darkgrey', linewidth = 5)
plt.title("Favorite Type of Movie", y=1.4, fontsize=20, color = 'Green',
         fontstyle = 'italic')
plt.pie(votes, labels=movie,
       labeldistance = 1.1,
       explode = [0.05,0.1,0.1,0],
       startangle = 90, 
        colors = ['lightgreen','red','pink','lightgrey'],
       autopct = "%0.2f%%",
       pctdistance=0.7,
       radius = 2,
       textprops={'fontsize' : 12},
       wedgeprops = {'edgecolor': 'purple', 
                     'linewidth' : 3, 'linestyle' : ':'},
       shadow = True)
centre_circle = plt.Circle((0,0),1.0,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(loc='upper right', bbox_to_anchor=(0,0.5,1.5,1), frameon = True)
plt.savefig("Financial Project Pie Graph")
plt.show()
```
