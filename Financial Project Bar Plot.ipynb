```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
grades = ['A', 'B', 'C', 'D']
test1 = [3,13,10,4]
test2 = [5,14,6,5]
test3 = [7,15,5,3]
```

## 1. **Draw The Bar Graph**

```python
plt.bar(grades, test1)
plt.show()
```

## 2. **Create Title, X and Y Label**

```python
plt.bar(grades, test1)
plt.title("Financial Project Bar Graph")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.show()
```

## 3. **Change Font Size of Title and X and Y Labels**

```python
plt.bar(grades, test1)
plt.title("Financial Project Bar Graph", fontsize = 18)
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("Number of Students", fontsize = 14)
plt.show()
```

## 4. **Set the Width and Align the Bars to the Center**

```python
plt.bar(grades, test1, width = 0.3, align = 'edge')
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.show()
```

```python
plt.bar(grades, test1, width = 0.3, align = 'edge',
        color = ['y','orange','g','red'], edgecolor = 'b')
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.show()
```

## 6. **Change Line Styles & Line Width of the Bars** 

```python
plt.bar(grades, test1, width = 0.3, align = 'edge',
        color = ['y','orange','g','red'], edgecolor = 'b', 
        linestyle = ':', lw = '3')
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.show()
```

## 7. Draw Bar Plots for Test1 and Test2

## 8. **Draw Bar Plots with Test1, Test2 & Test3**

```python
plt.bar(index, test1, width = 0.3, align = 'edge',
        color = 'y', edgecolor = 'g', 
        linestyle = ':', lw = '3')
plt.bar(index + 0.3, test2, width=0.3, align = 'edge', 
       color = 'red', edgecolor = 'g',
       ls=':', lw='3')
plt.bar(index + 0.6, test2, width=0.3, align = 'edge', 
       color = 'b', edgecolor = 'g',
       ls=':', lw='3')
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.xticks(index+0.45, grades)
plt.show()
```

## 9. **Add Legend**

```python
plt.bar(index, test1, width = 0.3, align = 'edge',
        color = ['y','b','red','black'], edgecolor = 'g', 
        linestyle = ':', lw = '3', label = 'Test1')
plt.bar(index + 0.3, test2, width=0.3, align = 'edge', 
       color = ['red','orange','g','b'], edgecolor = 'g',
       ls=':', lw='3', label = 'Test2')
plt.bar(index + 0.6, test2, width=0.3, align = 'edge', 
       color = ['b','g','purple','pink'], edgecolor = 'g',
       ls=':', lw='3', label = "Test3")
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.xticks(index+0.45, grades, rotation = 50)
plt.legend(loc= 'upper right', bbox_to_anchor=(0.5,0.5, 0.75, 0.5), frameon=True)
plt.show()
```

## 10. **Change Figure Size & Color**

```python
plt.figure(figsize=(10,5), facecolor = 'lightgreen', edgecolor = 'black', linewidth=3)
plt.bar(index, test1, width = 0.3, align = 'edge',
        color = 'y', edgecolor = 'g', 
        linestyle = ':', lw = '3', label = 'Test1')
plt.bar(index + 0.3, test2, width=0.3, align = 'edge', 
       color = 'red', edgecolor = 'g',
       ls=':', lw='3', label = 'Test2')
plt.bar(index + 0.6, test2, width=0.3, align = 'edge', 
       color = 'b', edgecolor = 'g',
       ls=':', lw='3', label = "Test3")
plt.title("Financial Project Bar Graph", fontsize =18, style = 'oblique')
plt.xlabel("Grades", fontsize = 14)
plt.ylabel("No of Students", fontsize = 14)
plt.xticks(index+0.45, grades, rotation = 50)
plt.legend(loc= 'upper right', bbox_to_anchor=(0.5,0.5, 0.75, 0.5), frameon=True)
plt.savefig('Financial Project Bar Plot')
plt.show()
```
